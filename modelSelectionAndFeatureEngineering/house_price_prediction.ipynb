```python
X = df.drop(columns=['price'])
y = df['price']

X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
LRegression = LinearRegression()
LRegression.fit(X_train, y_train) 
```

```python
#cofficidnets and intercept
print('Coefficients:', LRegression.coef_)
print('Intercept:', LRegression.intercept_)
```

```python
y_pred_train = LRegression.predict(X_train)
y_pred_test = LRegression.predict(X_test)
```

```python
#mad
from sklearn.metrics import mean_absolute_error
mad_train = mean_absolute_error(y_train, y_pred_train)
mad_test = mean_absolute_error(y_test, y_pred_test) 
print('Mean Absolute Deviation (Train):', mad_train)
print('Mean Absolute Deviation (Test):', mad_test)
```

```python
# Mean absoulte percentage error
print('Mean Absolute Percentage Error (Train):', np.mean(np.abs((y_train - y_pred_train) / y_train)) * 100)
print('Mean Absolute Percentage Error (Test):', np.mean(np.abs((y_test - y_pred_test) / y_test)) * 100)
```

```python
# rmse
from sklearn.metrics import mean_squared_error
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print('Root Mean Squared Error (Train):', rmse_train)
print('Root Mean Squared Error (Test):', rmse_test)
```
